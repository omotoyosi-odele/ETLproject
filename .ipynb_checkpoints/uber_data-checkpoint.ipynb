```python
import pandas as pd
from config import api_key
import pandas as pd
import numpy as np
import requests
import json

cab_rides = pd.read_csv("cab_rides.csv/cab_rides.csv")
cab_rides.head()
```

```python
uber_rides_data = cab_rides.loc[cab_rides["cab_type"] == "Uber"]
uber_rides_data.head()
```

```python
uber_rides_data["name"].unique()
```

```python
uber_rides = uber_rides_data[["distance", "source", "destination", "price", "time_stamp", "name"]]
uber_rides
```

```python
uber_rides_data["destination"].nunique()
```

```python
uber_rides_data["source"].nunique()
```

```python
destination = uber_rides_data["destination"]
```

```python
source = uber_rides_data["source"]

location = pd.concat([source, destination])
location
```

```python
uber_map = pd.DataFrame(location.unique())
#uber_map = pd.DataFrame(uber_rides_data["source"].unique())
```

```python
uber_map_source = uber_map.rename(columns={0: "source"})
uber_map_source["source_lat"] = ""
uber_map_source["source_lng"] = ""
uber_map_destination = uber_map.rename(columns={0: "destination"})
uber_map_destination["destination_lat"] = ""
uber_map_destination["destination_lng"] = ""
uber_map_destination['destination'] = uber_map_source['source']
uber_map_destination
```

```python
uber_map_source
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"
params = {
    "key": api_key,
    "inputtype" : "textquery",
    "fields": "geometry",
    "locationbias": "circle:50@42.3601,-71.0589"
}

for index, row in uber_map_source.iterrows():
    input = row["source"]
    
    params["input"] = input
    
    response = requests.get(base_url,params=params).json()
    results = response["candidates"]

    
    try:
        print(f"Grabbing where lats and longs for {row['source']}.")
    
        uber_map_source.loc[index,"source_lat"] =  results[0]['geometry']['location']['lat']
        uber_map_source.loc[index,"source_lng"] =  results[0]['geometry']['location']['lng']

        
    except(KeyError, IndexError):
        print("Missing field/result... skipping.")

    print("------------")
```
